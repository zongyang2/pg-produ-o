```python
# TESTE
# BASE 3: CALCULO DE PERDAS E ATRIBUIÇÃO DOS RISCOS DE PERDA
# TRATAMENTO DE DADOS: XXXXXX

import pandas as pd

# Leitura do arquivo CSV
df = pd.read_csv('/content/dados_final_64_produtos.csv')

# Selecionando até a coluna "Status consumo"
coluna_status_consumo_index = df.columns.get_loc('Status consumo')
df_filtrado = df.iloc[:, :coluna_status_consumo_index + 1]

# Garantindo que as colunas estejam no formato correto
df_filtrado['Data do vencimento'] = pd.to_datetime(df_filtrado['Data do vencimento'], errors='coerce')

# Função para limpar a coluna e garantir a conversão para numérico
def clean_numeric_column(column):
    # Removendo qualquer caractere não numérico (como vírgulas) e convertendo para float
    return pd.to_numeric(column.astype(str).str.replace('[^0-9.-]', '', regex=True), errors='coerce')

# Limpando as colunas numéricas
df_filtrado['Utilização livre'] = clean_numeric_column(df_filtrado['Utilização livre']) / 1000  # Dividindo por 1000
df_filtrado['Antes'] = clean_numeric_column(df_filtrado['Antes'])
df_filtrado['Consumo mínimo'] = clean_numeric_column(df_filtrado['Consumo mínimo'])
df_filtrado['Valor total'] = clean_numeric_column(df_filtrado['Valor total'])
df_filtrado['Estoque total'] = clean_numeric_column(df_filtrado['Estoque total']) / 1000  # Dividindo 'Estoque total' por 1000

# Substituindo valores ausentes (NaN) por 0
df_filtrado.fillna({'Utilização livre': 0, 'Antes': 0, 'Consumo mínimo': 0, 'Valor total': 0, 'Estoque total': 0}, inplace=True)

# Definindo a data fixa para cálculo de "Dias p/ vencer"
data_fixa = pd.to_datetime('2021-10-20')
df_filtrado['Dias p/ vencer'] = (df_filtrado['Data do vencimento'] - data_fixa).dt.days

# Adicionando a coluna "Perda com mínimo"
df_filtrado['Perda com mínimo'] = (
    (df_filtrado['Utilização livre'] + df_filtrado['Antes']) -
    (df_filtrado['Dias p/ vencer'] * df_filtrado['Consumo mínimo'] / 30)
)

# Aplicando a condição para que valores negativos sejam substituídos por vazio ("")
df_filtrado['Perda com mínimo'] = df_filtrado['Perda com mínimo'].apply(lambda x: '' if x < 0 else x)

# Adicionando a coluna "valor_PM"
df_filtrado['valor_PM'] = df_filtrado.apply(
    lambda row: (row['Valor total'] / row['Estoque total']) * row['Perda com mínimo'] if row['Perda com mínimo'] != '' else '',
    axis=1
)

# Salvando a base filtrada em um novo arquivo CSV
df_filtrado.to_csv('/content/dados_final_64_produtos_v3.csv', index=False)
# Exibindo o DataFrame no final
colunas_desejadas = ['Nº do material', 'Centro', 'Descrição centro', 'Data do vencimento',
                     'Utilização livre', 'Valor utiz. livre', 'Estoque total', 'Valor total',
                     'Prioridade', 'Antes', 'Dias p/ vencer', 'Consumo mínimo', 'Consumo médio',
                     'Consumo máximo', 'Consumo recente', 'Consumo MTD', 'Status consumo',
                     'Perda com mínimo', 'valor_PM']

# Exibindo o DataFrame com as colunas selecionadas
df_filtrado_selecionado = df_filtrado[colunas_desejadas]
print(df_filtrado_selecionado)



```

```python
# BASE XX: QUADRO RESUMO DO MAPEAMENTO POR RISCOS
# TRATAMENTO DE DADOS: XXXXXX


import pandas as pd

# Leitura do arquivo CSV
df = pd.read_csv('/content/base_tratada_v4.csv')

df['R$'] = (
    df['R$']
    .str.replace('.', '', regex=False)  # Remove os pontos (separadores de milhares)
    .str.replace(',', '.', regex=False)  # Substitui vírgulas por pontos (casas decimais)
    .astype(float)  # Converte para float
)

# Criar a tabela de resumo com pivot_table
tabela_resumo = pd.pivot_table(
    df,
    values='R$',                          # Coluna de valores
    index=['Localidade', 'Tipo do Centro'], # Índices: Tipo e Localidade
    columns='Risco',                      # Coluna do eixo X
    aggfunc='sum',                        # Soma os valores
    fill_value=0                          # Preencher valores ausentes com 0
)
# Adicionar a linha de TOTAL (somar todas as linhas)
tabela_resumo.loc['TOTAL'] = tabela_resumo.sum()
colunas_ordenadas = ['Altíssimo', 'Alto', 'Médio', 'Baixo', 'Baíxissimo', 'Já Vencido']
tabela_resumo = tabela_resumo[colunas_ordenadas]

# Exibir a tabela final
print(tabela_resumo)
```
